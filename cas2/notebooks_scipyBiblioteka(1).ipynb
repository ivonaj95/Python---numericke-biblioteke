# SciPy #
Scipy je nadogradnja Numpy biblioteke, što znači da sadrži  dodatne funkcionalnosti za naučno računanje. Scipy ima dodatne biblioteke za numeričku integraciju, optimizaciju, obradu signala, statistiku i drugo. Takođe, Scipy ima efikasnije implementacije nekih funkcija koje postoje i u Numpy-u, što znači da su često brže.

```python
import numpy as np
import matplotlib.pyplot as plt
# integracija
from scipy import integrate
from scipy import optimize
from scipy import interpolate
from scipy import linalg
```

```python
# racunanje odredjenog integrala funkcije sin(x) na intervalu [0, pi/2]
rezultat, greska = integrate.quad(np.sin, 0, np.pi/2)
print(f"Rezultat je {rezultat} sa tacnoscu {greska}")
```

```python
# pronalazenje minimuma i maksimuma funkcije x^2 + sin(5x)
def f(x):
    return np.sin(5 * x)
def g(x):
    return -f(x)
```

```python
# prvo cemo da nacrtamo grafik funkcije na intervalu [-100,100]
X = np.linspace(-np.pi/2,np.pi/2, 100)
plt.plot(X, f(X))

# trazimo minimu u okolini tacke 1
x0 = 1
# minimizacija funkcije f(x)
minimum = optimize.minimize(f, x0)

# ispis minimuma i tačke u kojoj je postignut minimum
print("Minimum je:", minimum.fun)
print("Tačka minimuma je:", minimum.x)

plt.scatter(minimum.x, minimum.fun)

# trazimo maksimum u okolini tacke 0.5 
# koristimo ponovo minimize metodu, ali ovaj put prosledjujemo -f funkciju
x1 = 0.5
maksimum = optimize.minimize(g, x1)
# ispis maksimuma i tačke u kojoj je postignut maksimum
print("Maksimum je:", -maksimum.fun)
print("Tačka maksimuma je:", maksimum.x)

plt.scatter(maksimum.x, -maksimum.fun)
plt.show()
```

```python
# interpolacija funkcije sin(x)
v = np.linspace(0, 10, 20)
u = np.sin(v)
plt.scatter(v, u)

f_interpolate = interpolate.interp1d(v, u)

# Kreiramo niz tacaka na kojima zelimo da izracunamo interpolovane vrednosti
vNovi = np.linspace(0, 10, 100)

plt.plot(vNovi, f_interpolate(vNovi))
plt.plot(vNovi, np.sin(vNovi))

plt.legend(["prava vrednost", "interpolacija", "sin(x)"])
```

```python
A = np.array([[1,2,3],[2,2,2],[3,1,2]])
print("determinanta matrice A", linalg.det(A))
print("inverz matrice A", linalg.inv(A))
Q, R = linalg.qr(A)
print(f"QR dekompozicija: \n ortogonalna matrica =\n {Q},\n gornje trougaona =\n {R}")
P,L,U = linalg.lu(A)
print(f"LU dekompozicija: \n donje trougaona = \n{L},\n gornje trougaona =\n {U}, \n matrica A dobijena posle gausovih transformacija\n {P}")
```
